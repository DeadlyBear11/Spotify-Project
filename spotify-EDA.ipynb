# Understanding Variables' Tags

### According to [Spotify's website](https://developer.spotify.com/documentation/web-api/reference/#/operations/get-audio-features)
* __Acousticness:__ 1.0 represents high confidence the track is acoustic. Range: {0.0,1.0}.
* __Analysis_url:__ A URL to access the full audio analysis of this track.
* __AlbumName:__ The title of the album.
* __AlbumId:__ The Spotify ID of the album.
* __ArtistName:__ The name of the main artist.
* __Danceability:__ Describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. Range {0.0,1.0}.
* __DateTime:__ I found no info about this variable. It seems to be related to the day I received my personal data from Spotify.
* __Duration_ms:__ The duration of the track in miliseconds.
* __EndTime:__ I found no info about this variable. It seems to be related to the day I received my personal data from Spotify.
* __Energy:__ Represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. Range {0.0,1.0}.
* __Id:__ The Spotify ID for the track.
* __Instrumentalness:__ Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". Range: {0.0,1.0}.
* __Key:__ The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1. Range: {-1,11}.
* __Liveness:__ Detects the presence of an audience in the recording. Range: {0.0,1.0}.
* __Loudness:__ The overall loudness of a track in decibels. Range {-60, 0}.
* __Mode:__ Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0. Range {0,1}.
* __MsPlayed:__ I found no info about this variable. It seems to be measuring how many milisenconds I played from a song, with a maximum value of the entire length of the song.
* __Name:__ A combination of the track title and the artist's name.
* __Speechiness:__ Detects the presence of spoken words in a track. Range {0.0,1.0}
* __Tempo:__ The overall estimated tempo of a track in beats per minute.
* __Time_signature:__ An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). Range: {3,7}.
* __Track_href:__ A link to the Web API endpoint providing full details of the track.
* __TrackName:__ The title of the track.
* __Type:__ The object type. All of them are "Audio_features".
* __Unnamed:__ A residual index.
* __Uri:__ The Spotify URI for the track.
* __Valence:__ Describes the musical emotional positiveness conveyed by a track. Range{0.0,1.0}.

# Data Visualization

```python
# Show me the 20 most played songs by me
top_tracks = df['trackName'].value_counts().head(20).to_frame()
top_tracks
```

```python
# Format Top 20 most played songs
top_tracks.reset_index(inplace=True)
top_tracks.rename(columns={'index':'Song_title', 'trackName':'Times_played'}, inplace=True)
top_tracks
```

```python
plt.figure(figsize = ( 10 , 6 ))
sns.barplot(x = 'Times_played', y = 'Song_title', data = top_tracks, palette= 'viridis_r')
plt.title( "Top 20 Most Played Songs" , size = 24 )
plt.xlabel("Number of times played", size = 16)
plt.ylabel("Song Title", size = 16)
```

```python
# Show me the 20 top artists for me
top_artists = df['artistName'].value_counts().head(20).to_frame()
top_artists
```

```python
# Format Top 20 artists for me
top_artists.reset_index(inplace=True)
top_artists.rename(columns={'index':'Artist_name', 'artistName':'Times_played'}, inplace=True)
top_artists
```

```python
plt.figure(figsize = ( 10 , 6 ))
sns.barplot(x = 'Times_played', y = 'Artist_name', data = top_artists, palette= "cividis_r")
plt.title( "Top 20 Most-listened-to Artists" , size = 24 )
plt.xlabel("Number of times played", size = 16)
plt.ylabel("Artist's Name", size = 16)
```
